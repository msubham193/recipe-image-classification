```python
from PIL import Image
import os
```

```python
for folder in os.listdir("dataset/"):
         if os.path.basename(folder)!=".DS_Store":
            for image in os.listdir("dataset/"+folder):
                if image.endswith(".gif") or image.endswith(".jpg") or image.endswith(".png") or image.endswith(".webp") or image.endswith(".jpeg"):
                    file_path = os.path.join("dataset/"+folder,image)
                    convert_to_jpg(file_path,folder)
```

```python
def convert_to_jpg(image_path, output_directory):
    try:
        # Open the image file
        image = Image.open(image_path)

        # Convert the image to JPG format
        image = image.convert("RGB")

        # Extract the filename and extension
        filename, extension = os.path.splitext(os.path.basename(image_path))

        # Create the output directory if it doesn't exist
        os.makedirs(output_directory, exist_ok=True)

        # Save the image as JPG in the output directory
        output_path = os.path.join(output_directory, f"{filename}.jpg")
        image.save(output_path, "JPEG")

        print(f"Image saved as JPG: {output_path}")
    except IOError:
        print(f"Unable to open or convert image: {image_path}")
```
